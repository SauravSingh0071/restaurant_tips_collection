```python
# Descriptive Statistics

print("\n--- Descriptive Statistics ---")
mean_tip = data['tip'].mean()
median_tip = data['tip'].median()
mode_tip = data['tip'].mode()[0]
std_tip = data['tip'].std()


print(f"Mean: {mean_tip}")
print(f"Median: {median_tip}")
print(f"Mode: {mode_tip}")
print(f"Standard Deviation: {std_tip}")
```

```python
# Load the data into a DataFrame
df = pd.DataFrame(data)

# Display the first few rows to inspect the original data
print("Original Data :-")
print(df.head())

# Create mapping dictionaries for each of the columns
sex_mapping = {
    'Female': 1,
    'Male': 2
}

smoker_mapping = {
    'No': 0,
    'Yes': 1
}

day_mapping = {
    'Sun': 7,
    'Mon': 1,
    'Tue': 2,
    'Wed': 3,
    'Thur': 4,
    'Fri': 5,
    'Sat': 6
}

time_mapping = {
    'Dinner': 1,
    'Lunch': 2
}

# Apply the mappings using map() for each column
df['sex'] = df['sex'].map(sex_mapping)
df['smoker'] = df['smoker'].map(smoker_mapping)
df['day'] = df['day'].map(day_mapping)
df['time'] = df['time'].map(time_mapping)

# Display the updated DataFrame to verify the changes
print("\nUpdated Data with Numeric Columns :-")
print(df.head())
```
